```python
import json
from datetime import datetime
import pandas as pd
df=pd.DataFrame({"Name":[""],"Email":[""],"Phone":[""],"Pdt.Id":[""],"Pdt. Type":[""],"Pdt. Name":[""],
                     "Quantity":[""],"Amount":[""],"Date":[""],"Time":[""]})

# records={1001:{"type":"chocolate","item":"5star","price":10,"quantity":350},
# 1002:{"type":"biscuit","item":"oreo","price":30,"quantity":450},
# 1003:{"type":"cake","item":"butter scotch","price":10,"quantity":570},
# 1004:{"type":"ice-cream","item":"vanilla","price":40,"quantity":130},
# 1005:{"type":"chips","item":"lays","price":20,"quantity":260}}

fd=open("Records.json","r")
js=fd.read()
records=json.loads(js)
fd.close()

#print(records)
print("-------------------MENU---------------------")
for prod in records:
    print("Pdt Id-",prod,":",records[prod]["type"],"|",records[prod]["item"],"|",records[prod]["price"],"|",records[prod]["quantity"])
print("-"*50+"\n")

name=input("Enter your name : ")
email=input("Entre your email : ")
phone=input("Enter you phone number :")

pr_id=input("Entre your product Id : ")
if pr_id not in records:
    print("Error!! You have entered wrong product id")
    exit()
pr_qn=int(input("How many quantity u wanna purchase :"))


if records[pr_id]["quantity"]>=pr_qn:

    print("------------------------------------------")
    print("           BILL                     ")
    print("Product Id      :",pr_id)
    print("Type            :",records[pr_id]["type"])
    print("Name            :",records[pr_id]["item"])
    print("Rate(Rs)        :",records[pr_id]["price"])
    print("Quantity        :",pr_qn)
    print("------------------------------------------")
    amt=records[pr_id]["price"]*pr_qn
    print("Billing Amount  :",amt)
    print("------------------------------------------\n")
    records[pr_id]["quantity"] = records[pr_id]["quantity"] - pr_qn

    #gettting our new transaction id from the last transaction.
    fid=open("id.txt","r")
    id=str(fid.read())
    fid.close()
    
    
    fd=open("Sales.csv","a+")
    cur_date = datetime.now().strftime("%d-%m-%Y")
    cur_time = datetime.now().strftime("%H:%M:%S")

    data= "\n"+id+","+name+","+ email +","+phone+","+ pr_id +","+records[pr_id]["type"]+","+ records[pr_id]["item"] +","+ str(pr_qn)+","+ str(amt)+","+ str(cur_date) +","+str(cur_time)
    print(data)
    fd.write(data)
    
    fid=open("id.txt","w")
    fid.write(str(int(id)+1))
    fid.close()
    fd.close()

elif records[pr_id]["quantity"]==0:
    print("OOPS Sorry !!! , Product is out of stock !!")

else:
    print("Sorry , we don't have enough quantity due to high demand")
    print("we are just left with",records[pr_id]["quantity"],"quantity")
    choice=input("Do you want to continue with this much ?? Y/N : ")
    if choice in ("Y","y"):
        print("-------------------------------------------")
        print("           BILL                     ")
        print("Product Id      :", pr_id)
        print("Type            :", records[pr_id]["type"])
        print("Name            :", records[pr_id]["item"])
        print("Rate(Rs)        :", records[pr_id]["type"])
        print("Quantity        :", records[pr_id]["quantity"])
        print("--------------------------------------------")
        print("Billing Amount  :", records[pr_id]["price"] * records[pr_id]["quantity"])
        print("-------------------------------------------\n")
        records[pr_id]["quantity"]=0
        
        #gettting our new transaction id from the last transaction.
        fid=open("id.txt","r+")
        id=str(fid.read())
        fid.close()

        fd=open("Sales.csv","a+")
        cur_date = datetime.now().strftime("%d-%m-%Y")
        cur_time = datetime.now().strftime("%H:%M:%S")

        data= "\n"+id+","+name+","+ email +","+phone+","+ pr_id +","+records[pr_id]["type"]+","+ records[pr_id]["item"] +","+ str(records[pr_id]["quantity"])+","+ str(amt)+","+ str(cur_date) +","+str(cur_time)
        print(data)
        fd.write(data)
        
        fid=open("id.txt","w")
        fid.write(str(int(id)+1))
        fid.close()
        fd.close()
        
    else:
        print("Thanks for showing interest. Bye Bye!!")
        exit()

#updating  data to json file

js=json.dumps(records)
fd=open("Records.json","w")
fd.write(js)
fd.close()
print("----------------------------------------------")
print("THANK YOU FOR PURCHASING, INVENTORY UPDATED")
print("----------------------------------------------")
```
